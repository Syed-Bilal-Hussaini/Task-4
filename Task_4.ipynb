```python
import pandas as pd
import numpy as np
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status', 'occupation', 'relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income']

data = pd.read_csv("/content/adult.csv",names= column_names , sep = ',',skipinitialspace=True)
numerical_featues = data.select_dtypes(include=['int64','float64']).columns.tolist()
categorical_features = data.select_dtypes(include=['object']).columns.tolist()
print(f"Numerical: {numerical_featues}")
print(categorical_features)
data.head()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import LabelEncoder

# Encoding the target variable 'income' (<=50K is 0, >50K is 1)
le = LabelEncoder()
data['income'] = le.fit_transform(data['income'])
```

```python
data = pd.get_dummies(data, columns=['workclass', 'marital_status', 'occupation',
                                 'relationship', 'race', 'sex', 'native-country'],
                    drop_first=True)

# Note: We drop 'education' because 'education-num' already provides the numeric rank
data.drop('education', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

numerical_features = ['age', 'fnlwgt', 'education_num', 'capital-gain', 'capital-loss', 'hours-per-week']

# Ensure these columns are of numeric type, coercing errors to NaN
# This is crucial because pd.read_csv might have inferred them as objects.
for col in numerical_features:
    if col in data.columns:
        data[col] = pd.to_numeric(data[col], errors='coerce')

# Drop rows where critical numerical features might have become NaN after coercion
# This handles any non-numeric entries that were converted to NaN.
data.dropna(subset=numerical_features, inplace=True)

# Now, scale the identified numerical features
data[numerical_features] = scaler.fit_transform(data[numerical_features])
```

```python
data.to_csv('adult_income_processed.csv', index=False)
print("Dataset saved successfully!")
```
